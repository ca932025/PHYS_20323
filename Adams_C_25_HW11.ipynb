```python
import numpy as np
# Copy your Particle class here
class Particle(object):
    def __init__(self, mass, x, y, z, vx, vy, vz):
        self.mass = mass # kg
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx # m/s
        self.vy = vy # m/s
        self.vz = vz # m/s
    def kinetic_energy(self): #(1/2)mv^2
        ke = (0.5) * self.mass * ( self.vx**2 + self.vy**2 + self.vz**2 )
        return(ke)
    def dist_from(self, other_particle):
        dist = np.sqrt( ((other_particle.x - self.x)**2) + ((other_particle.y - self.y)**2) + ((other_particle.z - self.z)**2) )
        return(dist)
    def update_pos(self, dt):
        # literally just multiply the velocity in that axis by the time interval
        self.x = self.x + self.vx * dt
        self.y = self.y + self.vy * dt
        self.z = self.z + self.vz * dt

# part 5
class ChargedParticle(Particle):
    
    def __init__(self, mass, x, y, z, vx, vy, vz, charge):
    
        super().__init__(mass, x, y, z, vx, vy, vz)
        # Add a new attribute for electric charge
        self.charge = charge

    # other_particle has to be a charged particle
    def find_seperation(self, other_particle):
        # first check if the other_particle is even a charged particle by checking if
        # it has the attribute "charge"
        if not hasattr(other_particle,"charge"):
            print("No Interaction, other_particle is not a charged particle")
            return
            
        print("Checking Seperation")
        print(self.dist_from(other_particle))
        if self.dist_from(other_particle) < 0.25:
            print("Interaction!")
            # check charge differences
            # if the charges are opposite
            if (self.charge > 0 and other_particle.charge < 0) or (self.charge < 0 and other_particle.charge > 0):
                # combine
                # set velocity of each to 0
                self.vx = 0.0
                self.vy = 0.0
                self.vz = 0.0
                other_particle.vx = 0.0
                other_particle.vy = 0.0
                other_particle.vz = 0.0
                # set charge of each to 0
                self.charge = 0.0
                other_particle.charge = 0.0
                print("Merge Interaction")
            elif self.charge == 0.0 and other_particle.charge == 0.0:
                # reverse velocity of each particle
                self.vx  *= (-1 * self.charge+other_particle.charge * (self.mass/other_particle.mass))
                self.vy  *= (-1 * self.charge+other_particle.charge * (self.mass/other_particle.mass))
                self.vz  *= (-1 * self.charge+other_particle.charge * (self.mass/other_particle.mass))
                other_particle.vx *= (-1 * self.charge+other_particle.charge * (other_particle.mass/self.mass))
                other_particle.vy *= (-1 * self.charge+other_particle.charge * (other_particle.mass/self.mass))
                other_particle.vz *= (-1 * self.charge+other_particle.charge * (other_particle.mass/self.mass))
                print("Repel Interaction")
            else:
                print("No Interaction")
                
        else:
            print("No Interaction")
            
```
